```python
import asyncio
from react_agent.llms.init_llms import OpenAILLM, GeminiLLM
from react_agent.core.react_agent import ReactAgent
from react_agent.tools.calculator import CalculatorTool
from react_agent.tools.web_search import SearchTool
from react_agent.core.memory import Memory
from dotenv import load_dotenv
load_dotenv()

import nest_asyncio
nest_asyncio.apply()

llm=GeminiLLM(
            api_key="",
            model="",
        )

async def main():
    agent = ReactAgent(
        llm=llm,
        tools=[SearchTool(api_key="Tavily_API_Key"), CalculatorTool()],
        memory=Memory(),
        verbose=True
    )

    result = await agent.run("who scored the most goals in world cup history?")
asyncio.run(main())
```
